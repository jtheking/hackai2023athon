```python
#delete all the space on all file name in the folder
import os
import re
import shutil
import glob
import sys
import time

def delete_space(path):
    for filename in os.listdir(path):
        char = '-'
        if filename.find(char) != -1:
            newname = filename.replace(char, '')
            os.rename(os.path.join(path, filename), os.path.join(path, newname))
            print('rename ' + filename + ' to ' + newname)
        else:
            print('no space in ' + filename)

delete_space('fridgeDatasetYOLO/train/images')
```
