```python
%cd C:\Users\danng\Documents\GitHub\hackai2023athon\computerVision
import torch
import utils

from IPython.display import Image, display
import requests
import numpy as np
import pandas as pd
import cv2
import mss
import json
import os

# Set the coordinates of the region of the screen to capture on the second monitor
monitor = {"top": 250, "left": 265, "width": 405, "height": 615}

# Create a directory to store the captured images
if not os.path.exists("captures"):
    os.makedirs("captures")

#array of all the items in the fridge
fridgeItems = ['avocado','beef','bell pepper', 'broccoli', 'butter', 'cabbage', 'carrot', 'cheese', 'chicken', 'egg','lettuce', 'milk', 'onion', 'pickle', 'tomato', 'zucchini']

# Initialize the mss (Multiple Screen Shots) module
with mss.mss() as sct:
    # Create a named window
    cv2.namedWindow("Fridge", cv2.WINDOW_NORMAL)

    # Loop over frames
    while True:
        # Get a screenshot of the region of the screen
        img = np.array(sct.grab(monitor))

        # Show the image
        cv2.imshow("Fridge", img)

        # Check if 's' key is pressed
        if cv2.waitKey(1) & 0xFF == ord('s'):
            print('s pressed: Closed Fridge')
            # Save the image to a file in the captures directory
            number = len(os.listdir("captures")) + 1
            filename = "captures/capture_{}.png".format(number)
            output = "output" + str(len(os.listdir("captures")) + 1)
            cv2.imwrite(filename, img)
            %cd yolov5
            yolov5FilePath = "../"+filename
            !python detect.py --weights best.pt --img 640 --conf 0.4 --source {yolov5FilePath} --save-txt --name {output} --hide-conf --line-thickness 1
            outputImage = Image(filename="runs/detect/{}/capture_{}.png".format(output,number))
            outputLabels = pd.read_csv("runs/detect/{}/labels/capture_{}.txt".format(output,number), sep=" ", header=None)
            outputLabels.columns = ["class", "x", "y", "width", "height"]
            #save class labels as a list
            classLabels = outputLabels["class"].tolist()
            outputArrays = []
            for i in classLabels:
                outputArrays.append(fridgeItems[i])
                # print(i, fridgeItems)
                # print(fridgeItems[classLabels[i]])

            %cd ..
            # with open(filename, 'rb') as f:
            #     print("yess")
            #     files = {'image': f}
            url = 'https://hackai2023athon-production.up.railway.app/set_ingredients'
            response = requests.post(url, data=json.dumps(outputArrays))
            print(response.text)
            print('\n')
            display(outputImage)
            print('\n')



        # Check if 'q' key is pressed
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

# Release resources
cv2.destroyAllWindows()
```
